```python
#merge 1st 2 sheets

df2 = pd.merge(df['main'],df['contacts'],on = ['uen', 'entity_name'], how = 'outer')
df2.head(5)
```

```python
df2.shape
```

```python
#filter criteria entity_status = 'Live', 'Live Company', email_check = 'Valid', designation != Manager, Executive 

df3 = df2[((df2['entity_status'] == 'Live') | (df2['entity_status'] == 'Live Company')) & (df2['email_check'] == 'Valid') & (df2['designation'] != 'Manager') & (df2['designation'] != 'Executive')]
df3
```

```python
#SSIC = target_ssic

df_tssic = df['target_ssic'].rename(columns={'target_ssic':'ssic'})
df4 = pd.merge(df3,df_tssic,on = 'ssic', how = 'inner')
df4
```

```python
df4['contact_status'].value_counts()
```

```python
#split contact status into 3 dataframes

df_notcon = df4[df4['contact_status'] == 'Not Contacted']
df_prevcon = df4[df4['contact_status'] == 'Previously Contacted']
df_repcon = df4[df4['contact_status'] == 'Repeat']
```

```python
#Randomise contact status - 'Not Contacted'

df_notconshuf = df_notcon.sample(frac=1)
df_notconshuf
```

```python
#Split contact_status 'Not Contacted' equally into 3 parts

df_notconshuf1 = df_notconshuf.iloc[:48]
df_notconshuf2 = df_notconshuf.iloc[48:97]
df_notconshuf3 = df_notconshuf.iloc[97:]
df_notconshuf2.shape
```

```python
#Randomise contact status - 'Previously Contacted'

df_prevconshuf = df_prevcon.sample(frac=1)
df_prevconshuf
```

```python
#Split contact_status 'Previously Contacted' equally into 3 parts

df_prevconshuf1 = df_prevconshuf.iloc[:46]
df_prevconshuf2 = df_prevconshuf.iloc[46:92]
df_prevconshuf3 = df_prevconshuf.iloc[92:]
df_prevconshuf1.shape
```

```python
#Randomise contact status - 'Repeat'

df_repconshuf = df_repcon.sample(frac=1)
df_repconshuf
```

```python
#Split contact_status 'Repeat' equally into 3 parts

df_repconshuf1 = df_repconshuf.iloc[:45]
df_repconshuf2 = df_repconshuf.iloc[45:89]
df_repconshuf3 = df_repconshuf.iloc[89:]
df_repconshuf1.shape
```

```python
#Combine into 3 files for 3 executive

df_exec1 = pd.concat([df_notconshuf1,df_prevconshuf1,df_repconshuf1],ignore_index=True)
df_exec2 = pd.concat([df_notconshuf2,df_prevconshuf2,df_repconshuf2],ignore_index=True)
df_exec3 = pd.concat([df_notconshuf3,df_prevconshuf3,df_repconshuf3],ignore_index=True)
df_exec1
```

```python
#Select columns for output template

df_new1 = df_exec1[['uen','entity_name','name','designation','phone','email','email_check']]
df_new2 = df_exec2[['uen','entity_name','name','designation','phone','email','email_check']]
df_new3 = df_exec3[['uen','entity_name','name','designation','phone','email','email_check']]
df_new1
```

```python
#export to excel

with pd.ExcelWriter('task2_output.xlsx') as writer:
    df_new1.to_excel(writer, sheet_name='Tom', index=False)
    df_new2.to_excel(writer, sheet_name='Jerry', index=False)
    df_new3.to_excel(writer, sheet_name='Mary', index=False)
```
